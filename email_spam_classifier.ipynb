# **Introduction of Email Classification**

Spam email classification is the process of automatically categorizing incoming emails as either legitimate or spam (unsolicited or unwanted bulk emails). It plays a crucial role in email filtering, which aims to separate unwanted messages from legitimate ones and prevent them from reaching the recipient's inbox.

Email spam is a significant problem, causing inconvenience, wasting time, and posing security risks to email users. Spam emails often contain fraudulent or malicious content, such as phishing attempts, scams, malware, or unwanted advertising. Filtering out such messages is essential for ensuring a clean and efficient email experience.

Here's an overview of the impact of spam email classification on email filtering:

**1. Reducing inbox clutter**: By accurately classifying emails as spam, the classification system helps in reducing the amount of unwanted messages that reach users' inboxes. This allows users to focus on important emails and avoid sifting through irrelevant or potentially harmful content.

**2. Protecting users from phishing and scams**: Phishing emails are fraudulent messages designed to trick recipients into revealing sensitive information like passwords or financial details. Effective spam classification systems can identify and flag such emails, protecting users from falling victim to these scams.

**3. Enhancing security**: Spam emails often contain malicious attachments or links to websites hosting malware. By identifying and filtering out these emails, classification systems help prevent users from inadvertently downloading or accessing harmful content, thus enhancing email security.

**4. Improving productivity**: Constantly dealing with spam emails can be time-consuming and distracting. Effective classification and filtering systems minimize the time spent on managing unwanted messages, allowing users to focus on important tasks and improving productivity.

**5. Enhancing user experience**: By reducing the presence of spam emails in users' inboxes, email classification systems improve the overall email experience. Users are less likely to miss important messages, have their attention diverted by irrelevant content, or be exposed to harmful or offensive material.

Spam email classification techniques typically involve the use of machine learning algorithms. These algorithms are trained on large datasets containing labeled examples of spam and legitimate emails. Features such as email headers, content, sender information, and patterns are analyzed to develop models that can accurately classify new incoming emails.

Overall, spam email classification significantly contributes to email filtering by reducing clutter, protecting against threats, enhancing security, improving productivity, and enhancing the overall user experience.

# **Classification Algorithms**

There are several classification algorithms that can be used for spam email classification. Here are some commonly employed algorithms:

**1. Naive Bayes**

**2. Support Vector Machines (SVM)**

**3. Decision Trees**

**4. Random Forest** , etc


Naive Bayes is a suitable algorithm for email spam classification because Naive Bayes is a relatively simple algorithm that is computationally efficient and easy to implement. It performs well even with large datasets, making it suitable for real-time email classification.Email spam classification often involves high-dimensional data, where each word or feature in the email represents a separate dimension. Naive Bayes assumes independence between features, which allows it to handle the curse of dimensionality effectively. This assumption simplifies the computations and makes Naive Bayes efficient in high-dimensional spaces.Naive Bayes is particularly well-suited for text classification tasks, including spam email classification. It treats each word or feature in an email independently and calculates the probability of an email being spam or legitimate based on the occurrence of specific words or features. This makes it effective in analyzing the presence or absence of certain keywords or patterns commonly found in spam emails.Despite its simplicity, Naive Bayes often performs remarkably well in spam email classification tasks. It has been widely used and has shown good accuracy in identifying spam emails, especially when trained on a large and representative dataset. It can handle both binary and multiclass classification problems effectively.Naive Bayes can work well even with limited training data. It can estimate the probabilities required for classification based on a relatively small number of labeled examples, making it useful when extensive labeled datasets are not available.Naive Bayes supports incremental learning, which means it can adapt to new data and update its classification model in real-time. This is valuable in spam email classification, as spammers constantly change their tactics, and the classification system needs to evolve accordingly.

While Naive Bayes does make the assumption of independence between features, which may not hold true in some cases, it still performs remarkably well in practice. Its simplicity, efficiency, and ability to handle high-dimensional text data make it a popular choice for email spam classification, especially as a baseline algorithm for comparison with more complex methods.

# **3. Text Preprocessing**

a. Lower case

b. Tokenization
    
c. Removing special characters

d. Removing stop words

e. Stemming

```python
!pip install nltk
```

```python

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

nltk.download('all')

from nltk.corpus import stopwords
```

```python
en_stops = set(stopwords.words('english'))
print(en_stops)
nltk.download('punkt')
```

```python
def text_preprocessing(text):
    text = text.lower()
    text = nltk.tokenize.word_tokenize(text, language='english', preserve_line=False)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') :
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
            
    return " ".join(y)
```

```python
df['preprocessed_text'] = df['Message'].apply(text_preprocessing)
```

# **6. Model Building** using Naive Bayes

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
## Using Gaussian Naive Bayes

gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
print(recall_score(y_test,y_pred1))
print(f1_score(y_test,y_pred1))
```

```python
## Using Multinomial Naive Bayes

mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
print(recall_score(y_test,y_pred2))
print(f1_score(y_test,y_pred2))
```

```python
## Using Bernoulli Naive Bayes

bnb.fit(X_train,y_train)
y_pred3 = bnb.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
print(recall_score(y_test,y_pred3))
print(f1_score(y_test,y_pred3))
```

```python
clfs = {
    'GaussianNB' : gnb,
    'MultinomialNB' : mnb,
    'BernoulliNB': bnb
}
def train_classifier(clf,X_train,y_train,X_test,y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)

    return accuracy,precision,recall,f1
```

```python
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for name,clf in clfs.items():

    current_accuracy,current_precision,current_recall,current_f1_score = train_classifier(clf, X_train,y_train,X_test,y_test)

    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    recall_scores.append(current_recall)
    f1_scores.append(current_f1_score)
```

```python

performance_df = pd.DataFrame({'Classifier':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores, 'Recall':recall_scores, 'F1 Score':f1_scores}).sort_values('Precision',ascending=False)
performance_df
```
